## e) Check if your test-cases can detect error if findMax is implemented as follows

```python
def findMax(num1: int, num2: int, num3: int):
    max_val = 0
    if (num1 > num2) and (num1 > num3):
        max_val = num1
    if (num2 > num1) and (num2 > num3):
        max_val = num2
    if (num3 > num1) and (num3 > num2):
        max_val = num3
    return max_val
```

```python
def solveQuartic(a, b, c, x):
    if a == 0 and b == 0 and c == 0:
        return -1
    
    if a == 0 and b == 0:
        return 0
    
    if a == 0:
        y = -c / b
        if y < 0:
            return 0
        x[0] = math.sqrt(y)
        x[1] = -math.sqrt(y)
        return 2
    
    delta = b * b - 4 * a * c
    if delta < 0:
        return 0
    
    y1 = (-b + math.sqrt(delta)) / (2 * a)
    y2 = (-b - math.sqrt(delta)) / (2 * a)
    
    count = 0
    if y1 >= 0:
        x[count] = math.sqrt(y1)
        count += 1
        x[count] = -math.sqrt(y1)
        count += 1
    if y2 >= 0 and y2 != y1:
        x[count] = math.sqrt(y2)
        count += 1
        x[count] = -math.sqrt(y2)
        count += 1
    
    return count
```

```python
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
    
x = [0.0] * 4  # Mảng 2để chứa nghiệm, tối đa 4 nghiệm
n = solveQuartic(a, b, c, x)
    
if n == -1:
    print("Infinite solutions.")
elif n == 0:
    print("No solution.")
else:
    print(f"The equation has {n} real solution(s): ", end="")
    for i in range(n):
        print(x[i], end=" ")
```

```python
test_cases = [
    ((0,0,0), (-1, [])),
    ((0,0,5), (0, [])),
    ((0,2,-8), (2, [2.0,-2.0])),
    ((1,0,-1), (2, [1.0,-1.0])),
    ((1,0,1), (0, [])),
    ((1,-2,1), (2, [1.0,-1.0])),
    ((1,5,6), (0, [])),
    ((1,-5,4), (4, [1.0,-1.0,2.0,-2.0])),
    ((2,-3,1), (4, [1.0,-1.0,math.sqrt(0.5),-math.sqrt(0.5)])),
    ((1,2,1), (0, [])),
]
for i, (inp, expected) in enumerate(test_cases, 1):
    x = [0.0] * 4
    n = solveQuartic(*inp, x)  
    roots = x[:n]               
    roots_set = set([round(r,5) for r in roots])
    expected_set = set([round(r,5) for r in expected[1]])
    if n == expected[0] and roots_set == expected_set:
        print(f"Test {i}: PASSED")
    else:
        print(f"Test {i}: FAILED (got {n}, {roots}, expected {expected})")
```
